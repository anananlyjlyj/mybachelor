```python
import matplotlib.pyplot as plt
import math
import scipy.constants as sc
import numpy as np
from scipy.optimize import fmin
import scipy.integrate as intg

def L (k, m, rol):
    if m == 0:
        return 1
    elif m == 1:
        return 1 + k - rol
    else: 
        return ((2*m - 1 + k - rol) * L(k, m-1, rol) - (m - 1 + k) * L(k, m-2, rol))/m
```

```python
#aB = 4 * math.pi * sc.epsilon_0 * (sc.hbar **2) / sc.m_e /(sc.e**2)
    
def Rnl (Z, r, n, l):
    p = 2 * r * Z / n
    wurzel = math.sqrt(8*(Z**3)*math.factorial(n - l - 1) / (((n**3) * 2 * n * math.factorial(n + l))))
    return math.exp(-p/2) * (p**l) * L(2*l+1, n-l-1, p) * wurzel
```

```python
# Im Folgenden sind Z als 1 gesetzt

maxx = []
for n in range(5):
    for l in range(n):        
        fig, ax1 = plt.subplots()
        x = np.linspace(0, 25, 1000)
        y = []
        z = []
        for t in x:
            y.append(Rnl(1, t, n, l))
            z.append(Rnl(1, t, n, l)**2 * t**2)
        max_z = max(z)  # Find the maximum y value
        max_x = x[z.index(max_z)]  # Find the x value corresponding to the maximum y value
        if l == n-1:
            maxx.append(max_x)
        print("n =",n)
        print("l =",l)
        
        ax1.plot(x, y, 'b-')
        ax1.set_xlabel('r')
        # Make the y-axis label, ticks and tick labels match the line color.
        ax1.set_ylabel('R_nl', color='b')
        ax1.tick_params('y', colors='b')
        
        ax2 = ax1.twinx()
        ax2.plot(x, z, 'r-')
        ax2.set_ylabel('Aufenthaltswahrscheinlichkeit', color='r')
        ax2.tick_params('y', colors='r')
        
        plt.show()


# Aufgabe c)
print("n, r mit Pmax")        
for n in range(1, 5):
    print (n, maxx[n - 1])
print("Kugelschale [r, r+dr] mit der hchsten Aufenthaltswahrscheinlichkeit in Abhngigkeit von Hauptquatenzahl fr die jeweilige maximale Drehimpulszahl(n + 1): P = 1 / n^2")
```

```python
for n in range(5):
    for l in range(n):
        x = np.linspace(0, 25, 1000)
        z = []
        for t in x:
            z.append(Rnl(t, n, l)**2 * t**2)
        plt.plot(x, z, label='n = %d, l = %d' %(n, l))
plt.legend()
plt.show()
```
