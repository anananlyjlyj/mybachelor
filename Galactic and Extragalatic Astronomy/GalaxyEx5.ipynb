```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# from file description we know
columns = ['Star ID', 'M_I', 'M_V', 'P', 'sigma_P', 't_brightest', 'A_I', 'R_21', 'phi_21','R_31', 'phi_31']
def metalicity(p, phi_31):
    return -3.142-4.092*p+0.824*(phi_31+np.pi)
```

```python
blg_data = pd.read_csv("RRab_blg.dat",header=None,delim_whitespace=True)
p = blg_data[3].values
phi_31 = blg_data[10].values
indices = np.where(phi_31=='-')
p = np.delete(p, indices).astype(np.float)
phi_31 = np.delete(phi_31, indices).astype(np.float)
z_blg = metalicity(p, phi_31)
z_median = np.median(z_blg)
p_mean_blg = np.mean(p)
plt.style.use("fivethirtyeight")
plt.figure(figsize = [8, 8])
plt.hist(z_blg, bins = 20)
plt.title("RRab_blg metallicity")
plt.xlabel("Metallicity")
plt.ylabel("Number of Stars")
plt.show()
print("The mean period of the RR Lyrae stars that lie in the direction of BLG is:")
print("%0.3f days" % (p_mean_blg))
print("The median metallicity of the RR Lyrae stars that lie in the direction of BLG is:")
print("%0.2f +/- %0.2f dex" % (z_median, np.std(z_blg)))
print(len(p))
```
