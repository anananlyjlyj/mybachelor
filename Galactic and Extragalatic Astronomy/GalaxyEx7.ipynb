```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.constants as const


v_rot = [167.7, 132, 60.5, 133, 136.5]
log_v_rot = np.log10(v_rot)
log_L_B_L_sun = [10.184, 9.76, 8.376, 9.772, 9.82]
log_L_B = log_L_B_L_sun # + np.log10(const.L_sun.value)

plt.style.use("fivethirtyeight")
m,b = np.polyfit(log_v_rot, log_L_B, 1)
#plt.plot(log_v_rot, m*log_v_rot+b, "g", label="linear fit")
plt.scatter(log_v_rot, log_L_B, s = 20, label="from calculation")
plt.plot(log_v_rot, 3.42+3.09*log_v_rot, "r", label="from literature")
plt.xlabel("log(v_rot)")
plt.ylabel("log(L_B/L_sun)")
plt.legend()
plt.title("Tully-Fisher relation")
print(m)
print(b)
```
