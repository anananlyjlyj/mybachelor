```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.constants as const

mu_e = 20
n = 1

def mu(x):
    b_n = 1.9992*n-0.3271
    return mu_e + 2.5*b_n/np.log(10) * (x**(1/n)-1)

r = np.linspace(0, 8, 50)
plt.style.use("fivethirtyeight")
fig, ax = plt.subplots()
ax.plot(r, mu(r))
ax.axhline(23, color='red', lw=2, alpha=0.5, label="limiting surface brightness of image")
ax.axvline(2.65, color='green', lw=2, alpha=0.5, label = "maximal visible radius")
plt.title("Surface brightness versus radius")
plt.xlabel("R/R_e")
plt.ylabel("mu [mag/arcsec^2]")
plt.legend()
plt.show()
```
