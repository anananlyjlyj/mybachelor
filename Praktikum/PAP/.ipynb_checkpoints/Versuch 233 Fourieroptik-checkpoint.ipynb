```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Berechnung der Beugungsbilder

# Beugungsfunktion Spalt
def beugung_spalt(x):
    return np.sinc(x)**2

# Beugungsfunktion Doppelspalt
def beugung_doppelspalt(x):
    return np.sinc(x)**2 * np.cos(np.pi * v * x)**2

# Anzahl a der Nebenmaxima die dargestellt werden sollen:
a = 1
# Verhaeltnis Spaltabstand zu Spaltbreite:
v = 2
x = np.linspace(-(a+1), a+1, 200)
plt.plot(x, beugung_spalt(x), label = "Spalt")
plt.plot(x, beugung_doppelspalt(x), label = "Doppelspalt")
plt.xlabel("$\pi x$")
plt.ylabel("Intensitaet/b.E.")
plt.title("Beugung am Spalt und Doppelspalt")
plt.ylim((0, 1.1))
plt.legend()
plt.show()
```

```python
# Berechnung der modifizierten Spaltbilder

# Integrad
def spalt(k):
    return d/np.pi*np.sin(k*d/2)/(k*d/2)*np.cos(y*k)

# Doppelspalt
def doppelspalt(k):
    return 2 * d/np.pi * np.cos(k*g/2) * np.sin(k*d/2) / (k*d/2) * np.cos(y2*k)

d = 1 # Spaltbreite
n = 3 # Integrationsbereich
Y = np.linspace(-1, 1, 200)*d # y-Werte
# Spaltabstand in Einheiten der Spaltbreite
g = 2 * d
Y2 = np.linspace(-2, 2, 400)*d

from scipy.integrate import quad
f_modifiziert=[]
f2_modifiziert=[]
for y in Y:
    result, error = quad(spalt, 0, 2*np.pi*n/d)
    f_modifiziert.append(result**2)
for y2 in Y2:
    result2, error2 = quad(doppelspalt, 0, 2*np.pi*n/d)
    f2_modifiziert.append(result2**2)

# Kurve auf 1 nomieren
f_modifiziert /= np.max(f_modifiziert)
f2_modifiziert /= np.max(f2_modifiziert)

plt.plot(Y, f_modifiziert, label = "Spalt")
plt.plot(Y2, f2_modifiziert, label = "Doppelspalt")
plt.xlabel("y / d")
plt.ylabel("Intensitaet / b.E.")
plt.title("modifiziertes Objektbild Spalt und Doppelspalt")
plt.ylim((0, 1.1))
plt.legend()
plt.show()
```
