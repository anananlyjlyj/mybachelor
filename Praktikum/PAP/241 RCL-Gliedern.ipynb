```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
def ph(f,a):
    return 180./np.pi * np.arctan(1/(2*np.pi*f*a))

# 3. Phasengang
f = np.linspace(1000,10000,10)
x = np.linspace(1000,10000,100)
phi = np.array([68,58,49.7,44.6,34.2,28.9,28.2,24.8,23.3,19.4])
delta_phi=np.array([10,10,1.5,2.0,2.5,1.2,1.4,1.6,1.8,2.0])
popt,pcov = curve_fit(ph, f, phi)
plt.errorbar(f, phi, yerr = delta_phi, fmt = '.',label='Messung')
plt.plot(x, ph(x, popt[0]))
plt.title("Phasenverschiebung gegen Frequenz -- RC Glied")
plt.ylabel("Phasenverschiebung [Â°]")
plt.xlabel("Frequenz [Hz]")
plt.xscale('log')
plt.legend(loc='best')
#plt.savefig('2.pdf', format='pdf',pixel=1000)
plt.show()
print(popt[0],np.sqrt(pcov[0][0]))
```
