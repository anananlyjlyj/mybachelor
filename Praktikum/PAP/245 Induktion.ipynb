```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import matplotlib.style as mlps
mlps.use('classic')
# induzierte Spannung als Funktion von Frequenz
I = 4.014
fehler_I = 0.01
f = np.array([2.88, 6.04, 9.13, 12.0, 14.9])
fehler_f = np.array([0.02, 0.02, 0.04, 0.1, 0.2])
U_ind = np.array([2.08, 4.4, 6.6, 8.5, 10.8])/2
fehler_U_ind = np.array([0.02, 0.08, 0.2, 0.2, 0.2])
# Fit
def linear(x, c):
    return x * c
p0 = [0.7]
popt, pcov = curve_fit(linear, f, U_ind, p0)
# Plot
plt.errorbar(f, U_ind, xerr = fehler_f, yerr = fehler_U_ind, label = 'Datei', fmt = '.')
x = np.linspace(0, 16, 100)
plt.plot(x, linear(x, popt[0]), label = 'Fit')
plt.legend(loc = 'best')
plt.title("Abb.3.1: induzierte Spannung gegen Frequenz bei I = (4.014 $\pm$ 0.01)A")
plt.xlabel('f / Hz')
plt.ylabel('$U_ind$ / V')
plt.axis((0, 16, 0, 12))
plt.savefig('1.pdf', format = 'pdf', pixel=1000)
plt.show()
print('Die Steigung ist: ',round(popt[0],4),'+-',round(np.sqrt(pcov[0][0]),4))
```
