```python
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
a = 0.5
b = 8
ranx = []
rany = []
for i in range(1000):
    x = rd.uniform() * a
    y = rd.uniform() * a * b
    # if y < 8x, x accepted; if not, rejected
    if y < b * x: 
        ranx.append(x)
        rany.append(y)
plt.plot(ranx, rany)
p = np.linspace(0, 0.5, 1000)
plt.plot(p, b * p)
plt.show()
```

```python
N=[1e4, 5e4, 1e5, 5e5]

def distance(a, b):
    d = np.sqrt(a**2+b**2)
    return d

# accuracy
A = np.zeros(4)

for i in range(4):
    x=rd.uniform(-1,1,N[i])
    y=rd.uniform(-1,1,N[i])

    plt.scatter(x[(distance(x,y) < 1)], y[(distance(x,y) < 1)], color = 'blue')
    plt.scatter(x[(distance(x,y) > 1)], y[(distance(x,y) > 1)], color = 'red')
    plt.show()
    
    a=(len(y[(distance(x,y)<1)]))
    print(a/N[i]*4)
    A[i] = a/N[i]*4
    
plt.plot(N, np.array([abs(t - np.pi) for t in A]))
plt.show()
    
```
