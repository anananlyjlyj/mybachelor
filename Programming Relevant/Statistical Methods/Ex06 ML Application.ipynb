# Estimating a constant in the presence of Gaussian noise

K measurements: $x_k =c+\epsilon_k$ with noise $\epsilon_k$ ~ $ N(0, s_k^2)$, what is the MLE(maximum likelihood estimation) of c?

$L(c|x_k, s_k, (Gauss))=\prod_{i=1}^K {\frac{1}{\sqrt{2\pi}s_k}e^{-\frac{1}{2}\frac{(x_k-c)^2}{s_k^2}}}$

$l = \sum_{k=1}^K \{-\ln \sqrt{2\pi}-\ln{s_k}-\frac{1}{2}\frac{(x_k-c)^2}{s_k^2}\}$

$\frac{dl}{dc} = \sum \frac{x_k-c}{s_k^2}=0 \Rightarrow \hat{c}=\frac{\sum{\frac{x_k}{s_k^2}}}{\sum{\frac{1}{s_k^2}}}$

$\frac{d^2l}{dc^2} = \frac{1}{\sum{\frac{1}{s_k^2}}} \Rightarrow \sigma_c^2=\frac{1}{\sum{\frac{1}{s_k^2}}}$

if for every k $s_k = s$ then $\sigma_c^2=\frac{1}{\frac{k}{s_k^2}}$, that is $\sigma_c=\frac{s}{\sqrt k}$

# Entropy(isoperimetric problem)

p(x) with $x\in [0, \infty[$, $\int_0^{infty} {p(x)dx}=1$, $\int_0^{infty} {xp(x)dx}=\mu$

Extremalize $S=-\int_0^{\infty}{p(x)\ln(p(x))dx}$

Equal think in physics, total energy of a system is preserved, what is the maximal entropy?

Variation calculus

$L=F+\lambda G$ with G the integrand of the known constains

Language: $L(p, x, \mu)=-p(x)\ln p(x)+\lambda_1 p(x) + \lambda_2 xp(x)$

Euler equation: $\frac{d}{dx}\frac{\partial L}{\partial p'}-\frac{\partial L}{\partial p}=0 \Rightarrow \frac{\partial L}{\partial p}=-\ln p -1+\lambda_1+\lambda_2 x = 0 \Rightarrow p(x)=e^{\lambda_1-1}e^{\lambda_2 x}=ce^{\lambda_2 x}$

Set the formel of p(x) into two constrains: $\int_0^{\infty} {ce^{\lambda_2 x}dx}=1$$\Rightarrow \lambda_2 = -c$ and $\int_0^{\infty} {cxe^{\lambda_2 x}dx}=1$$\Rightarrow \lambda_2 = -\frac{1}{\mu}$

Thus we get $p(x)=\frac{1}{\mu}e^{-\frac{x}{\mu}}$
