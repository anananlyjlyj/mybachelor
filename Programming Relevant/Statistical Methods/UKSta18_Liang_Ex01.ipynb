# R exercise 05: control statements
We implement a function that prints out the first n(as argument of the function) prime numbers:

```r
Prime <- function(n){
    if(n < 1)
        return("Invalid argument") 
    result <- 2
    for (i in 3:(50*n)){
        if (any(i %% 2:(i/2) == 0))
            next
        result <- c(result, i)
        if (length(result) == n)
            return(result)
    }
}
# Test
print(Prime(100))
```
