```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math

data = pd.read_csv("Rotation_problem7.3.dat",header=None,skiprows=3, delim_whitespace=True)
x = data[0].values
y = data[1].values
v_r = data[2].values
d_v_r = data[3].values
x_med = np.median(x)
y_med = np.median(y)
v_r_med = np.median(v_r)
v_r_dev = v_r - v_r_med

angle = 48
length = 1.5
endy = length * math.sin(math.radians(angle))
endx = length * math.cos(math.radians(angle))

plt.style.use("fivethirtyeight")
plt.figure(figsize = [8, 8])
plt.scatter(x[v_r_dev < 0], y[v_r_dev < 0], s=-2*v_r_dev[v_r_dev < 0], label="stars with v_r < v_r_med")
plt.scatter(x[v_r_dev > 0], y[v_r_dev > 0], s=2*v_r_dev[v_r_dev > 0], label="stars with v_r > v_r_med")
plt.scatter(x_med, y_med, s=10, label="(Median) Center of system")
plt.plot([x_med-endx, x_med+endx], [y_med-endy, y_med+endy], "g",linewidth = 1, label = "angle = 48Â°")
plt.xlim(min(x)-0.5, max(x)+0.5)
plt.ylim(min(y)-0.5, max(y)+0.5)
plt.title("Position of stars")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import numpy.linalg

d = np.abs(endy*2*x-endx*2*y+(x_med+endx)*(y_med-endy)-(y_med+endy)*(x_med-endx))/(2*length)
rot = (v_r - v_r_med)/d
print(rot)
plt.scatter(d[np.abs(rot) < 2], rot[np.abs(rot) < 2])
plt.show()
```
